```python
from material import *
from glass import *
from model import *
import matplotlib.pyplot as plt

# libraries
material_lib = MaterialLibrary("./dat/COATING.dat") 
glass_lib = GlassLibrary("./agf/OHARA.agf")

# substrate and film material
SBSL7 = glass_lib.find_glass("S-BSL7")
SIO2 = material_lib.find_material("SIO2_EM")
TA2O5 = material_lib.find_material("TA2O5_EM")
AL2O3 = material_lib.find_material("AL2O3_EM")

# layer stack model
AR_400_700_3L = LayerStackModel()
AR_400_700_3L.substrate.material = SBSL7
AR_400_700_3L.set_layers([[SIO2, 0.087221100520488],
                         [TA2O5, 0.118906265728823],
                         [AL2O3, 7.65425558950326E-02]])
```

```python
# plot reflectance vs wavelength at 0 deg

AR_400_700_3L.incident_angle = 0.0
w_list = np.linspace(0.300, 0.800, 100)
Rs = []
Rp = []
for w in w_list:
    AR_400_700_3L.wavelength = w
    Rs.append(AR_400_700_3L.reflection("s"))
    Rp.append(AR_400_700_3L.reflection("p"))

plt.plot(w_list, Rs)
plt.plot(w_list, Rp)
plt.show()
```

```python
# plot reflectance vs wavelength at 30 deg

AR_400_700_3L.incident_angle = np.deg2rad(30.0)
w_list = np.linspace(0.300, 0.800, 100)
Rs = []
Rp = []
for w in w_list:
    AR_400_700_3L.wavelength = w
    Rs.append(AR_400_700_3L.reflection("s"))
    Rp.append(AR_400_700_3L.reflection("p"))

plt.plot(w_list, Rs)
plt.plot(w_list, Rp)
plt.show()
```

```python
# plot reflectance vs angle at 0.55 micron

AR_400_700_3L.wavelength = 0.55
angle_list = np.linspace(0.0, 90, 100)
Rs = []
Rp = []
for a in angle_list:
    AR_400_700_3L.incident_angle = np.deg2rad(a)
    Rs.append(AR_400_700_3L.reflection("s"))
    Rp.append(AR_400_700_3L.reflection("p"))

plt.plot(angle_list, Rs)
plt.plot(angle_list, Rp)
plt.show()
```
